```python
# -*- coding: utf-8 -*-
"""
Spyder Editor

This is a temporary script file.
"""


from random import random 

class Tree :
    k = 1  #  counter used to compute val
    total_nodes = 1   # number of nodes in the tree
    
    def __init__(self, max,  a) :
        #  particulier
        self.succ = None # next node
        self.val = 1 # task name
        self.level = 1 # node level
        self.leaf = False # true if leaf else false
        self.total_levels = 3 
        self.pred=None
        self.time = int((random() * max + 1))   # execution time of the task
    def generate(self, root,  nb_pred_max,  nb_lev,  max_time,  index) :
        # there are nodes which have no predecessor
        print("val of the root is   " + str(self.k))
        print("adress of the root is   " + str(root))
        root.total_levels = nb_lev
        # n = 0 --> number of the predecessors of root
        if (root.level < nb_lev) :
            if ((root.level == 1)) :
                n = int((random()* nb_pred_max + 1))
            else :#  level of the node is different of 1
                n = int((random() * nb_pred_max + 1))
                #  number of predecessors between 0 and maxpred
                # n=n-1;
                #  n=1;
                if (n == 0) :
                    root.leaf = True
            self.settotal_nodes ( self.total_nodes + n)
            #update the number of nodes in the tree
            root.pred = [None] * (n) 
            i = 0
            while (i < len(root.pred)) :
                print("The number of nodes is now " + str(self.total_nodes))
                self.k += 1
                # to ensure distinct spots are obtained
                root.pred[i] = Tree(max_time, i)
                root.pred[i].level = root.level + 1
                # racine.pred[i].nb_total_niv=nb_niveaux;
                #each node in the tree contains an attribute indicating the number of levels in the tree
                root.pred[i].succ = root
                #  if(racine.pred[i].succ!=null)
                root.pred[i].val = self.k
                print("The node number   " + str(i) + "  at level   " + str(root.pred[i].level) + "  its value=    " + str(root.pred[i].val) + "  its successor adress is   " + str(root.pred[i].succ))
                print("*********************************************")
                self.generate(root.pred[i], nb_pred_max, nb_lev, max_time, i)
                i += 1
        else :
            root.leaf = True
            root.pred=None # to define the end of each branch
            
        print("The attribute of root leaf is " + str(root.leaf))
        #print(root)
    def  gettotal_nodes(self) :
        return  self.total_nodes
    def settotal_nodes(self, a) :
        self.total_nodes = a
    def  get_node_val(self) :
        return  self.val
    # Special function to print tree items
    def print_level(self,n,level_no=0):
        #to print level X items
        if(n.pred!=None): # Continue while there are more items 
            for i in n.pred: # loop with items that have the same level 
                self.print_level(i,level_no)
        if level_no==0:
            print("L= ",n.level,", Val=",n.val)
        elif(n.level==level_no):
            print(n.val,end='\n')
            
            
class Datascience :
    def main(self, args) :
        N = Tree(5, 0)
        N.generate(N, 3, 3, 10, 0)
        # root name ** nb_predecessors max ** nb_level ** maximum execution time ** index
        print("++++++++++++++++++")
        print("the total number of nodes in the tree is " + str(N.gettotal_nodes()))

        try:
            print("the 4 in tree is " + str(N.pred[0].pred[2].get_node_val()))
        except Exception as e:
            print("No Node at this index\n",type(e))
            print("Total_levels: ",N.total_levels)
            print("Total_nodes: ",N.total_nodes)
        print("print all items at level 4:")
        N.print_level(N,4) # print all items at level 4
        print("print all items of the tree:")
        N.print_level(N,) # print all items of the tree
#main function of the class
if __name__=="__main__":
    Datascience().main([])
```
